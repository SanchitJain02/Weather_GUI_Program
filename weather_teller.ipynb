```python
from tkinter import *
import tkinter as tk
from geopy.geocoders import Nominatim
from tkinter import ttk,messagebox
from timezonefinder import TimezoneFinder
from datetime import datetime
import requests
import pytz
import time
```

```python
root = Tk()
root.title("Weather App")
root.geometry("1200x600+300+200")
root.resizable(False,False)


def getWeather():
    try:
        city = textfield.get()
        geolocator = Nominatim(user_agent = "geoapiExcercises")
        location = geolocator.geocode(city)
        obj = TimezoneFinder()
        result = obj.timezone_at(lng = location.longitude,lat = location.latitude)
        print(result)

        home = pytz.timezone(result)
        local_time = datetime.now(home)
        current_time = local_time.strftime("%I:%M %p")
        clock.config(text = current_time)
        name.config(text = "CURRENT WEATHER")

        api = "https://api.openweathermap.org/data/2.5/weather?q="+city+"&appid=e9aac0e9fa9b421ee549eef46273e23f"
        json_data = requests.get(api).json()
        condition = json_data["weather"][0]["main"]
        description = json_data["weather"][0]["description"]
        temp = int(json_data["main"]["temp"] - 273.15)
        min_temp = int(json_data["main"]["temp_min"] - 273.15)
        max_temp = int(json_data["main"]["temp_max"] - 273.15)
        pressure = json_data["main"]["pressure"]
        humidity = json_data["main"]["humidity"]
        wind = json_data["wind"]["speed"]
        sunrise = time.strftime("%I:%M:%S",time.gmtime(json_data["sys"]["sunrise"] - 19800))
        sunset = time.strftime("%I:%M:%S",time.gmtime(json_data["sys"]["sunset"] - 19800))

        t.config(text =(temp,"째"))
        c.config(text = (condition,"|","FEELS","LIKE",temp,"째"))

        w.config(text = wind)
        h.config(text = humidity)
        d.config(text = description)
        p.config(text = pressure)
        mat.config(text = (max_temp,"째"))
        mit.config(text = (min_temp,"째"))
        s_rise.config(text = sunrise)
        s_set.config(text = sunset)
    
    except Exception as e:
        messagebox.showerror("Weather App","Invalid Entry!!!")
        


    
Search_image = PhotoImage(file = "search.png")
myimage = Label(image = Search_image)
myimage.place(x = 20,y=20)

textfield = tk.Entry(root,justify = "center",width = 17,font = ("poopins",25,"bold"),bg = "#404040",border = 0,fg = "white")
textfield.place(x = 50,y = 40)
textfield.focus()

Search_icon = PhotoImage(file = "search_icon.png")
myimage_icon = Button(image = Search_icon,borderwidth = 0,cursor = "hand2",bg = "#404040",command = getWeather)
myimage_icon.place(x = 400,y = 34)

Logo_image = PhotoImage(file = "logo_2.png")
logo = Label(image = Logo_image)
logo.place(x = 250,y = 100)

Frame_image = PhotoImage(file = "frame1.png")
frame_myimage = Label(image = Frame_image)
frame_myimage.place(x = 70,y = 350)


name = Label(root,font = ("arial",15,"bold"))
name.place(x = 30,y = 100)
clock = Label(root,font = ("Helvetica",20))
clock.place(x = 30,y = 130)

label1 = Label(root,text = "WIND",font=("Helvetica",15,"bold"),fg = "white",bg = "#157DEC")
label1.place(x = 120,y = 400)

label2 = Label(root,text = "HUMIDITY",font=("Helvetica",15,"bold"),fg = "white",bg = "#157DEC")
label2.place(x = 225,y = 400)

label3 = Label(root,text = "DESCRIPTION",font=("Helvetica",15,"bold"),fg = "white",bg = "#157DEC")
label3.place(x = 430,y = 400)

label4 = Label(root,text = "PRESSURE",font=("Helvetica",15,"bold"),fg = "white",bg = "#157DEC")
label4.place(x = 650,y = 400)

label5 = Label(root,text = "MAX. TEMP",font=("Helvetica",15,"bold"),fg = "white",bg = "#157DEC")
label5.place(x = 120,y = 500)

label6 = Label(root,text = "MIN. TEMP",font=("Helvetica",15,"bold"),fg = "white",bg = "#157DEC")
label6.place(x = 270,y = 500)

label7 = Label(root,text = "SUNRISE",font=("Helvetica",15,"bold"),fg = "white",bg = "#157DEC")
label7.place(x = 425,y = 500)

label8 = Label(root,text = "SUNSET",font=("Helvetica",15,"bold"),fg = "white",bg = "#157DEC")
label8.place(x = 600,y = 500)

t = Label(font = ("arial",100,"bold"),fg = "#ee666d")
t.place(x = 500,y = 140)
c = Label(font = ("arial",15,"bold"))
c.place(x = 500,y = 290)

w = Label(text = "...",font = ("arial",20,"bold"),bg = "#157DEC")
w.place(x = 120,y = 430)
h = Label(text = "...",font = ("arial",20,"bold"),bg = "#157DEC")
h.place(x = 280,y = 430)
d = Label(text = "...",font = ("arial",20,"bold"),bg = "#157DEC")
d.place(x = 435,y = 430)
p = Label(text = "...",font = ("arial",20,"bold"),bg = "#157DEC")
p.place(x = 670,y = 430)
mat = Label(text = "...",font = ("arial",20,"bold"),bg = "#157DEC")
mat.place(x = 120,y = 530)
mit = Label(text = "...",font = ("arial",20,"bold"),bg = "#157DEC")
mit.place(x = 270,y = 530)
s_rise = Label(text = "...",font = ("arial",20,"bold"),bg = "#157DEC")
s_rise.place(x = 425,y = 530)
s_set = Label(text = "...",font = ("arial",20,"bold"),bg = "#157DEC")
s_set.place(x = 600,y = 530)



root.mainloop()
```
